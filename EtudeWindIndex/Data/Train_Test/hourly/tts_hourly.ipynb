# Imports

```python
import pandas as pd
from pathlib import Path
from sklearn.model_selection import train_test_split
```

# Fonction de création de set

```python
def real_tts(project):
    
    #Lecture de l'export 10min du projet
    S3_bucket = 'edfred-edfre-sbx-eu-west-1-solar-radiation-data'
    S3_real_folder = r'EtudeWindIndex/Real'
    file = '10min_' + project
    S3_real = f's3://{S3_bucket}/{S3_real_folder}/{file}.csv'
    real_set = pd.read_csv(S3_real, sep=';')
    
    #Fabrication du trainset et testset
    #Stratifié par turbine et par heure !
    trainset, testset = train_test_split(real_set, test_size=0.33, random_state=42, stratify=real_set.loc[:,['turbine','year','month','day','hour']])
    
    #Sauvegarde
    trainset_path = Path('/home/ec2-user/SageMaker/EtudeWindIndex/Data/Train_Test/hourly/trainset.csv')
    testset_path = Path('/home/ec2-user/SageMaker/EtudeWindIndex/Data/Train_Test/hourly/testset.csv')
    trainset.to_csv(trainset_path, index=False, sep=';')
    testset.to_csv(testset_path, index=False, sep=';')
```
